```python
# top 10 most sold products (same thing as above)

fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_Category')['Orders'].sum().sort_values(ascending=False).plot(kind='bar')
```

```python
summary = df.groupby(['Product_Category', 'State'])['Orders'].sum().reset_index()

# Select the top 10 Product Categories
top_categories = summary.groupby('Product_Category')['Orders'].sum().nlargest(10).index

# Filter the summary DataFrame for only top categories
top_summary = summary[summary['Product_Category'].isin(top_categories)]

# Create a bar plot with hue for State
plt.figure(figsize=(12, 7))
sns.barplot(data=top_summary, x='Product_Category', y='Orders', hue='State', ci=None)

# Set the title and labels
plt.title('Top 10 Product Categories by Orders with Hue by State', fontsize=16)
plt.xlabel('Product Category', fontsize=14)
plt.ylabel('Total Orders', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='State')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
columns_to_drop = ['Cust_name', 'Product_ID', 'Status', 'unnamed1']

existing_columns_to_drop = [col for col in columns_to_drop if col in df.columns]
cleaned_data = df.drop(columns=existing_columns_to_drop).dropna(subset=['Amount'])

encoded_data = pd.get_dummies(cleaned_data, columns=['Gender', 'Age Group', 'State', 'Zone', 'Occupation', 'Product_Category'], drop_first=True)

X = encoded_data.drop(columns=['Amount'])
y = encoded_data['Amount']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = LinearRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mse ** 0.5

print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)

```
